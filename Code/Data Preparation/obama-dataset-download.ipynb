```python
    import os
    import subprocess
    from pytube import YouTube
    from moviepy.editor import VideoFileClip

    def download_video(url, output_path, filename):
        yt = YouTube(url)
        ys = yt.streams.get_highest_resolution()
        ys.download(output_path=output_path, filename=filename)

    def extract_audio(video_path, output_path):
        video_clip = VideoFileClip(video_path)
        audio_clip = video_clip.audio
        audio_clip.write_audiofile(output_path, )

    def split_frames(video_path, output_path='frames', bitrate='60k'):
        if not os.path.exists(output_path):
            os.makedirs(output_path)

        command = f'ffmpeg -i {video_path} -b:v {bitrate} -f image2 {output_path}/frame_%04d.jpg'
        subprocess.run(command, shell=True)

    def extract_structural_info(video_path):
        # Get motion vectors, motion residuals, coding block organization, and quantization parameters
        command = f'ffprobe -v error -select_streams v:0 -show_entries frame=pkt_pts_time,motion_vector,side_data_list -of default=noprint_wrappers=1 {video_path}'
        result = subprocess.run(command, shell=True, stdout=subprocess.PIPE, text=True)
        print(result.stdout)

    if __name__ == "__main__":
        f = open("/kaggle/input/obama-dataset/obama.txt", "r")
        youtube_urls = f.readlines()[101:201]
        for i in range(len(youtube_urls)):
            j = 101 + i
            download_video(youtube_urls[i],"videos", f"obama{j:03}.mp4")
            video_path = f'/kaggle/working/videos/obama{j:03}.mp4'
            extract_audio(video_path, f"/kaggle/working/audios/obama{j:03}.wav")
            print(f"data - {j} done")
    #         split_frames(video_path)
    #         extract_structural_info(video_path)

```
